```python
def sizegrp(row):
    if   (row['pct0']<=row['com_ret']<row['pct10']):
        group=1
    elif (row['pct10']<=row['com_ret']<row['pct20']):
        group=2
    elif (row['pct20']<=row['com_ret']<row['pct30']):
        group=3
    elif (row['pct30']<=row['com_ret']<row['pct40']):
        group=4
    elif (row['pct40']<=row['com_ret']<row['pct50']):
        group=5
    elif (row['pct50']<=row['com_ret']<row['pct60']):
        group=6
    elif (row['pct60']<=row['com_ret']<row['pct70']):
        group=7
    elif (row['pct70']<=row['com_ret']<row['pct80']):
        group=8
    elif (row['pct80']<=row['com_ret']<row['pct90']):
        group=9
    elif (row['pct90']<=row['com_ret']<=row['pct100']):
        group=10
    else:
        group=np.nan
    return pd.Series({'cusip':row['cusip'], 'yearmonth':row['yearmonth'],'com_ret': row['com_ret'],'REV6':row['REV6'] ,'group': group})
```
